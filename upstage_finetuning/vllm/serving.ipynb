```python
def test_generation(prompt, temp, max_tokens=100, top_p=0.9, top_k=20, min_p=0.1):
    response = client.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=temp,
        max_tokens=max_tokens,
        top_p=top_p,
        extra_body={ 
            'top_k':top_k,
            'min_p' : min_p
        }
    )
    return response.choices[0].message.content
```

```python
prompt="세상에서 제일 맛있는 음식은 뭐야?"
test_generation(prompt, 0.7, 200, 0.9, 20, 0.1)
```

```python
df = pd.DataFrame(data)
```

```python
def get_answer(text):
    try:
        res = client.chat.completions.create(
            model=model_name,
            messages=[
                {"role": "user", "content": text}
            ],
            temperature=1.2,
            max_tokens=500,
            top_p=0.7,
            extra_body={ 
                'top_k': 10,
                'min_p' : 0.2   
            }
        )
        return res.choices[0].message.content
    except Exception as e:
        return str(e)
```

```python
df['answer'] = df['input_text'].apply(get_answer)
```
