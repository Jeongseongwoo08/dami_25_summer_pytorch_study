```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "4"
import torch
```

```python
problems = ['Calculate this: (12345 * 6789) + (98765-43210) / 5 ',
            'Solve: 888.88*1234.45 - 678.90 / 2.5',
            'What is the result of: ((512 + 256) * 16) / (64-32)+128',
            'Compute: 999999 / 33+ 77777 * 3 - 11111',
            'Calculate: (45.6 * 7.8) + (90.1 / 0.5) - (12.3 * 4.5)',
            'Solve this equation: 10000 - (256 * 13) + (1024 / 8) * 5',
            'What is the value of: (1+2+3+4+5) * (100-55) / 9',
            'Calculate: 2468 * 13579 + 86420',
            'Compute: ((1200 / 25) *45)+ ((800/20) * 15)',
            'Solve: 5000+ 50 *50 -50 / 0.5']
```

```python
import time
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name='google/gemma-2-2b-it'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map='auto')
```

```python
from transformers import pipeline
```

```python
pipe = pipeline(
    'text-generation',
    model=model,
    tokenizer = tokenizer,
)
```

```python
time_start = time.time()

results = pipe(problems, max_new_tokens = 1024)

time_end = time.time()

print('Time takne:', time_end-time_start)
```

```python
# vllm으로 inference 측정
```

```python
import vllm
from vllm import LLM, SamplingParams
from typing import List, Optional
import time
from vllm.outputs import RequestOutput
```

```python
model_name='google/gemma-2-2b-it'

model=LLM(model=model_name, tensor_parallel_size=1, gpu_memory_utilization=0.9, max_model_len=1024)
```

```python
sampling_params = SamplingParams(max_tokens = 1024, temperature=1)
```

```python
start_time = time.time()

outputs: List[RequestOutput] = model.generate(problems, sampling_params)

end_time = time.time()

print('Time taken: ', end_time-start_time)
```
